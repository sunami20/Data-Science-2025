#Matplotlib

```python
from matplotlib import axes
#Creat 2x2 grid
fig, axes =plt.subplots(2,2, figsize=(10,10))

#Plot 1 : Histogram
axes[0, 0].hist(student['Age'], bins=20, color='green')
axes[0, 0].set_title('Age Score')

# Plot 2 : Bar Chart
student['gender'].value_counts().plot(kind='bar', ax=axes[0, 1])
axes[0, 1].set_title('Gender Count')

# Plot 3 : Scatter
axes[1, 0].scatter(student['id'][:20], student['Age'][:20])
axes[1, 0].set_title('Id vs Age')

# Plot 4 : Line
axes[1, 1].plot(student['birthday'][:5])
axes[1, 1].set_title('Birtday Score')

plt.tight_layout()
plt.show()
```

#Seaborn

```python
sns.histplot(student['id'])
plt.title('Id Score')
plt.show()

# with KDE curve

sns.histplot(student['id'], kde=True)
plt.title('Id Score with Kde')
plt.show()
```

#Box Plot

```python
# Gender vs Id Score
sns.boxplot(x='gender', y='id', data=student)
plt.grid(True, which="both", linestyle="--", linewidth=0.7, alpha=0.5)
plt.title('Id by Gender')
plt.show()
```

#Count Plot

```python
sns.countplot(x='visit_per_week', data=student)
plt.title('Count by visit_per_week')
plt.show()
```

```python
sns.barplot(x='visit_per_week', y= 'avg_time_in_gym', data=student)
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
sns.kdeplot(data=student, x='id', hue='gender', fill=True)
plt.grid(True, which="both", linestyle="--")
plt.title('Id score by Gender')
plt.show()
```

```python
attend_counts = student['attend_group_lesson'].value_counts()
plt.pie(attend_counts, labels=attend_counts.index, autopct='%1.2f%%')
plt.title('Attend Group Lesson')
plt.show()
```
