```python
#3: Conditiobal Expressions

num = int(input("Enter your number:"))
if num % 2==0:
  print("The number is even")
elif  num %2!=0:
  print("The number is odd")
else:
  print("The number is zero")
```

```python
#Assignment 2: Factorial of Number
def factorial(n):
  iniatial = 1
  for i in range (1, n+1):
      iniatial = iniatial *i
  return iniatial
result = factorial(5)
print(result)


```

```python
#Assignment 3: Multiplication Table
n=int(input("Enter ur number: "))
for i in range(1,n):
  for j in range (1,n):
    output= i*j
    print(f"{i} X {j} = ",{output})
  print("______")
```
