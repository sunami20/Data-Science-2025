# **Numpy**

```python
#Import necessary library
import numpy as np

```

```python
#creat a numpy array with 20 random integers between 1 and 100

arr = np.random.randint(1,100, 20)

#Display yhe Numpy array
print(" The Numpy array is: \n ", arr)
```

```python
#1. Find the sum of all elements
#The np.sum() function computes the sum of all elements in the array

sum = np.sum(arr)
print("The sum of arr \n", sum)
```

```python
#2. Find the maximum and minimum values in the array
#np.max() returns the maximum value and np.min() returns the minimum value in the array

max = np.max(arr)
min = np.min(arr)
print("The number of max", max)
print("The number of min", min)
```

# **Pandas**

```python
#Import necessary library
import pandas as pd
```

```python
#1. Creat a DatFrame with 5 rows and 3 columns of ranadom numbers
#The DataFrame is generated from a 5x3 maatrix of random integers between 1 and 100

data =(np.random.randint(1, 100, size=(5 , 3)))

df =pd.DataFrame(data, columns=['A', 'B', 'C'])
print(df)
```

```python
# 2. Compute the sum of each column
#The .sum() function computes the sum for each column in the DataFrame

sum_colum = df.sum()
print(sum_colum)
```

```python
# 3. Compute the mean of each column
# The .mean() function computes the mean (average) of each column in the DataFrame

mean_colum=df.mean()
print(mean_colum)
```
