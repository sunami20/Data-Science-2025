# **Function**

```python
#Task9 : Write a Python function that takes a string as input and prints the first and last character.
def first_last_char(text):
    if len(text) == 0:
        print("The string is empty.")
    else:
        print("First character:", text[0])
        print("Last character:", text[-1])


first_last_char("name")
```
